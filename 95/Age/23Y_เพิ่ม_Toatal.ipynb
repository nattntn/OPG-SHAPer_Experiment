```python
from itertools import combinations

# List of names
names = [
    'Lower Second Molar', 'Upper Second Premolar', 'Lower Second Premolar',
    'Upper Canine', 'Mandibular angle', 'Upper First Premolar',
    'Upper Lateral Incisor', 'Posterior border of Ramus', 'Upper Second Molar',
    'Lower First Molar', 'Lower First Premolar', 'Condyle', 'Upper First Molar',
    'Lower Third Molar', 'Lower Canine', 'Upper Central Incisor',
    'Upper Third Molar', 'Nasal'
]

# Get all combinations of names with length 2
pairs_of_names = list(combinations(names, 2))
print(pairs_of_names)
```

```python
len(pairs_of_names)
```

```python
import pandas as pd

# Dictionary to store results for each pair
pair_results = {}

for name1, name2 in pairs_of_names:
    # Identify filenames that contain name1
    A = set(age_23_neg[age_23_neg['name'] == name1]['filename'])

    # Identify filenames that contain name2
    B = set(age_23_neg[age_23_neg['name'] == name2]['filename'])

    # Find the intersection of the two sets to get filenames that contain both name1 and name2
    common_filenames = A.intersection(B)

    # Filter the DataFrame to include only the common filenames
    filtered_filenames = age_23_neg[age_23_neg['filename'].isin(common_filenames)]

    # Filter the DataFrame to include only the name1 or name2
    filtered_region = filtered_filenames[filtered_filenames['name'].isin([name1, name2])]

    # Dictionary to store name counts and associated filenames
    name_count_neg = {}
    name_files = {}

    # List of names
    list_neg = filtered_region['name'].tolist()
    # List of filenames
    filenames = filtered_region['filename'].tolist()

    # Iterate over each name and corresponding filename
    for name, filename in zip(list_neg, filenames):
        if name in name_count_neg:
            name_count_neg[name] += 1
            name_files[name].append(filename)
        else:
            name_count_neg[name] = 1
            name_files[name] = [filename]

    # Sort the name_count_neg dictionary by count of occurrences (descending order)
    sorted_names = sorted(name_count_neg.items(), key=lambda x: x[1], reverse=True)

    # Store the results in the dictionary
    pair_results[(name1, name2)] = {
        'sorted_names': sorted_names,
        'name_files': name_files
    }

    # Print a summary for this pair
    print(f"Pair: {name1} and {name2}")
    print("Unique names and their occurrences in negative SHAP values area:")
    total_unique_names = len(sorted_names)
    print(f"Total Unique Names: {total_unique_names}")

    for i, (name, count) in enumerate(sorted_names, start=1):
        unique_filenames = len(set(name_files[name]))  # Calculate unique filenames
        print(f"{i}. {name}: {count}/{len(filtered_region['filename'].unique())} ({(count / len(filtered_region['filename'].unique())):.2%}) of total")
        print(f"    Filenames: {name_files[name]}")

    print("Filenames with the same name:")
    for name, filenames in name_files.items():
        if len(set(filenames)) >= 1:  # Only consider names that appear in more than one filename
            print(f"{name}: {', '.join(filenames)}")
```

```python
import pandas as pd

# Dictionary to store results for each pair
pair_results = {}

for name1, name2 in pairs_of_names:
    # Identify filenames that contain name1
    A = set(age_23_pos[age_23_pos['name'] == name1]['filename'])

    # Identify filenames that contain name2
    B = set(age_23_pos[age_23_pos['name'] == name2]['filename'])

    # Find the intersection of the two sets to get filenames that contain both name1 and name2
    common_filenames = A.intersection(B)

    # Filter the DataFrame to include only the common filenames
    filtered_filenames = age_23_pos[age_23_pos['filename'].isin(common_filenames)]

    # Filter the DataFrame to include only the name1 or name2
    filtered_region = filtered_filenames[filtered_filenames['name'].isin([name1, name2])]

    # Dictionary to store name counts and associated filenames
    name_count_pos = {}
    name_files = {}

    # List of names
    list_pos = filtered_region['name'].tolist()
    # List of filenames
    filenames = filtered_region['filename'].tolist()

    # Iterate over each name and corresponding filename
    for name, filename in zip(list_pos, filenames):
        if name in name_count_pos:
            name_count_pos[name] += 1
            name_files[name].append(filename)
        else:
            name_count_pos[name] = 1
            name_files[name] = [filename]

    # Sort the name_count_pos dictionary by count of occurrences (descending order)
    sorted_names = sorted(name_count_pos.items(), key=lambda x: x[1], reverse=True)

    # Store the results in the dictionary
    pair_results[(name1, name2)] = {
        'sorted_names': sorted_names,
        'name_files': name_files
    }

    # Print a summary for this pair
    print(f"Pair: {name1} and {name2}")
    print("Unique names and their occurrences in posative SHAP values area:")
    total_unique_names = len(sorted_names)
    print(f"Total Unique Names: {total_unique_names}")

    for i, (name, count) in enumerate(sorted_names, start=1):
        unique_filenames = len(set(name_files[name]))  # Calculate unique filenames
        print(f"{i}. {name}: {count}/{len(filtered_region['filename'].unique())} ({(count / len(filtered_region['filename'].unique())):.2%}) of total")
        print(f"    Filenames: {name_files[name]}")

    print("Filenames with the same name:")
    for name, filenames in name_files.items():
        if len(set(filenames)) >= 1:  # Only consider names that appear in more than one filename
            print(f"{name}: {', '.join(filenames)}")
```

```python
# Optional: Create a DataFrame to keep all the information together
age_pos = []
for (name1, name2), result in pair_results.items():
    for name, count in result['sorted_names']:
      if name in name1:
        age_pos.append({
            'name': name,
            'pair': name2, #'pair': f"{name1} and {name2}"
            'count': count,
            'total_1': len(age_23_pos[age_23_pos['name']== name1]),
            'percentage_1':  f"{(count / len(age_23_pos[age_23_pos['name']== name1])) * 100 :.2f}%",
            'total_2': len(age_23_pos[age_23_pos['name']== name2]),
            'percentage_2':  f"{(count / len(age_23_pos[age_23_pos['name']== name2])) * 100 :.2f}%",
            'total_files': len(age_23_pos['filename'].unique()),
            'percentage_files':  f"{(count / len(age_23_pos['filename'].unique())) * 100 :.2f}%",
            'filenames': result['name_files'][name]
        })

pos_2 = pd.DataFrame(age_pos)
pos_2
```
