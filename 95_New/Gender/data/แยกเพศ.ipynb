## Positive Gender (937)

```python
test_gender_pos = pd.read_csv (r'/content/gdrive/My Drive/Tooth_Shap_GPT/Deep_tooth/OPG_SHAPer/Result_ALL/final_result_gender_positive_bboxes.csv')
test_gender_pos = test_gender_pos.iloc[:,1:]
print(test_gender_pos.shape)
test_gender_pos
```

```python
len(set(test_gender_pos['filename']))
```

### ทายเพศถูก (818)

```python
# เลือกเฉพาะ row ที่ทายถูก
c = test_gender_pos[((test_gender_pos['Gender_predict_str'] == 'Female') & (test_gender_pos['Gender'] == 'F')) | ((test_gender_pos['Gender_predict_str'] == 'Male') & (test_gender_pos['Gender'] == 'M'))]
c
```

```python
len(set(c['filename']))
```

```python
# ทำค่า Prob
c['prob'] = c.apply(lambda row: 1 - row['Gender_predict'] if row['Gender'] == 'F' else row['Gender_predict'], axis=1)
```

```python
c
```

```python
table = c.reset_index()
table = table.iloc[:,1:]
table
```

```python
table.to_csv('/content/gdrive/My Drive/Tooth_Shap_GPT/Deep_tooth/OPG_SHAPer/Result_ALL/Gender/data/gender_pos_correct.csv')
```
